Step 8: Find out the outliers from the dataset according to the columns and fill them with the mean.

```python
# column 'sale_price'
Q1 = df['sale_price'].quantile(0.25)
Q3 = df['sale_price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Replace outliers with mean
df['sale_price'] = np.where(
    (df['sale_price'] < lower_bound) | (df['sale_price'] > upper_bound),df['sale_price'].mean(),df['sale_price'])
```

```python
# column 'market_price'
Q1 = df['market_price'].quantile(0.25)
Q3 = df['market_price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound_m = Q1 - 1.5 * IQR
upper_bound_m = Q3 + 1.5 * IQR

# Replace outliers with mean
df['market_price'] = np.where(
    (df['market_price'] < lower_bound) | (df['market_price'] > upper_bound),df['market_price'].mean(),df['market_price'])
```

Step 9: Create Plots or visualizations.


```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set figure size
sns.set(rc={"figure.figsize": (20, 7)})

# Check if 'category' exists and is not empty
if 'category' in df.columns and not df['category'].isnull().all():
    # Ensure 'category' is treated as a categorical column
    df['category'] = df['category'].astype('category')
    
    # Create the countplot with explicit data parameter
    ax = sns.countplot(data=df, x="category")
    
    # Add count labels to the bars
    for bars in ax.containers:
        ax.bar_label(bars, fmt='%d')

    # Rotate x-axis labels for readability
    plt.xticks(rotation=90)

    # Add title and labels
    plt.title("Count of Each Category", fontsize=16)
    plt.xlabel("Category", fontsize=14)
    plt.ylabel("Count", fontsize=14)

    # Show the plot
    plt.show()
else:
    print("Column 'category' is missing or has no valid data!")
```
